```python
#필요한 회사만 추출한다
final_df = pd.DataFrame(columns = ['사업장명', '업종','연금보험료', '직원수', '신규취득자수', '상실가입자수', ' 적용일자'])
list_id = ["쏘카", "비바리퍼블리카", "무신사", "크래프톤",'야놀자','에이프로젠','지피클럽']
for name in list_id:
    new_df = df[df['사업장명'].str.contains(name)]
    final_df = pd.concat([final_df, new_df])
```

```python
final_df = final_df.drop(index=final_df.query('연금보험료 == 0').index)
```

```python
final_df["년도"] = "2021"
```

```python
final_df.sort_values(by="연금보험료", ascending=False,inplace=True)
```
